##SOLUCION-DFS

```python
dictionary = {}
for i in PROVINCIA:
  a = PROVINCIA[i]['Origen']
  dictionary[a] =[]

for i in PROVINCIA:
  a = PROVINCIA[i]['Origen']
  b = PROVINCIA[i]['Destino']
  c = PROVINCIA[i]['Distancia'] 
  midict = {b:c}
  dictionary[a].append(midict)

dictionary
```

```python
#Listas de adyacencia para cada nodo
adj = [[] for _ in range(len(dictionary))]

i = -1
vis = []
orden = []

for origen in dictionary.keys():
  a = origen
  orden.append(a)
  i = i + 1
  for destino_dict in dictionary[origen]:
    
    for z,key in enumerate(destino_dict):
     
      b = key
      c = destino_dict[key]
      adj[i].append((b,c))

adj
```

```python
def valid(nodo:str):
  if nodo in vis:
    return False
  else:
    return True
```

```python
import heapq as pq #priority queue

arr = []
def DFS(nodito: str, peso: float):
  vis.append(nodito)

  index = list(orden).index(nodito)
  #Busco su index para recorrer sus vecinos

  q = [] # q es la cola soporte de heapq
  for e, w in adj[index]:
    pq.heappush(q, (w, e))
  
  while len(q) > 0 :
    n,v = pq.heappop(q)
    if valid(v):
      arr.append((nodito,v,n))
      DFS(v, n)
      break
```

#Solución -Dijkstra

```python
import heapq as pq #priority queue
import math

#Listas de adyacencia para cada nodo
adj = [[] for _ in range(len(dictionary))]
i = -1
vis = {}
dis = {}
path = {}
for origen in dictionary.keys():
  #print(origen)
  a = origen
  vis[a] = False
  dis[a] = math.inf
  path[a] = '-1'
  i = i + 1

  for destino_dict in dictionary[origen]:
    #print(destino_dict)
    for z,key in enumerate(destino_dict):
      #print(key)
      #print(destino_dict[key])
      b = key
      c = destino_dict[key]
      adj[i].append((b,c))
  
adj
```

```python
def dijkstra(adj: list, start: str):
        
    q = [] # cola de prioridad
    dis[start] = 0

    # En nuestra cola vamos a llevar 
    # pares (w, v) = (weight, vertex) = (peso, vertice/nodo)
    # por qué no (v, w) -> las tuplas se ordenan 
    # lexicograficamente por defecto

    pq.heappush(q, (0, start)) 

    while len(q) > 0:
        _, v = pq.heappop(q)

        if vis[v]: # Si ya lo he visitado antes, no hago nada
            continue 

        vis[v] = True
        path 
        index = list(vis).index(v)
        for e, w in adj[index]: # Recorro vecinos
            # Si he encontrado una nueva ruta para
            # llegar al vecino 'e' a través de 'v' usando
            # la arista con peso 'w', meto esa opción
            # a la cola de prioridad
            
            if dis[v] + w < dis[e] and vis[e] == False:
                dis[e] = dis[v] + w
                path[e] = v
                pq.heappush(q, (dis[e], e))

    return dis
```

```python
import time
start_time = time.time()
dijkstra(adj,'PEDRO RUIZ')
print("--- %s seconds ---" % (time.time() - start_time))
```

```python
ans = dijkstra(adj, 'PEDRO RUIZ')
##Visitados
print(vis)
costo = 0
##Resultados
print('Path de la solución')
print(path)
for i in ans:
  costo+=ans[i]
  print('Distancia minima para llegar a '+ str(i) +' -> ' + str(ans[i]))
print('Costo total del recorrido '+str(costo))
```

#Solución - Brute Force

<h1>Fuerza Bruta</h1><br>
<b><i>Es una tecnica trivial pero a menudo usada, que consiste en enumerar todo los psibles candidatos para la solución de un problema. </i></b>
    <p>Recorremos distintos tipos de caminos necesitando  memoria constante de tal manera que comparemos y obtengamos el menor recorrido de entre todas las soluciones iteradas.</p>

##Algoritmo en general

 <p>Debido a que permutaremos la cantidad de posibilidades del recorrido, en busqueda del camino más corto, la complejidad de esta solución tendrá un tiempo de ejecución de factor polinómico de orden O((n − 1)!), el factorial del número de centros poblados para esta solución es impracticable.</p>

*   10 centro poblados = 9! = 362880  posibilidades
*   100 centro poblados = 99! = 9.332622 E+155 posibilidades
*   1000 centros poblados = 999! = 10^2565 posibilidades
*   Numero de Átamos en el Universo: <(10^100)
