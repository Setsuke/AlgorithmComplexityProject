```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('MatrizDeCentrosPobladoss.csv')

print(df.head())
```

```python
df2 = df[df['DISTRITO']=="JAZAN"]
print(df2)
```

```python
df2 = df2.reset_index(drop=True)
print(df2)
```

```python
df2 = df2[["CENTRO POBLADO", "LATITUD", "LONGITUD"]]
print(df2)
```

```python
import itertools
centros=df2['CENTRO POBLADO']
list(itertools.combinations(centros,2))
```

```python
latitudes=df2['LATITUD']
list(itertools.combinations(latitudes,2))
```

```python
longitudes=df2['LONGITUD']
list(itertools.combinations(latitudes,2))
```

```python
combinacionesCentros = list(itertools.combinations(centros,2))
combinacionesLatitudes = list(itertools.combinations(latitudes,2))
combinacionesLongitudes = list(itertools.combinations(longitudes,2))

df3 = pd.DataFrame(combinacionesCentros,columns=['Origen','Destino'])
df4 = pd.DataFrame(combinacionesLatitudes,columns=['LatitudO','LatitudD'])
df5 = pd.DataFrame(combinacionesLongitudes,columns=['LongitudO','LongitudD'])
```

```python
df6=df3.join(df4)
df6.join(df5)
df7=df6.join(df5)
```

```python
import math
def calcularDistancia(row):  
  distancia = math.sqrt((row["LatitudO"] - row["LatitudD"]) ** 2 + (row["LongitudO"] - row["LongitudD"]) ** 2)
  return distancia
```

```python
df7["Distancia"] = df7.apply(calcularDistancia,axis=1)
df7
```

```python
df8=df7[["Origen","Destino","Distancia"]]
```

```python
df8
```

```python
diccionario=df8.to_dict('index')
```
